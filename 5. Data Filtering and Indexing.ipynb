```python
import pandas as pd
```

```python
# Filtering based upon condition
df = pd.read_excel("Excel_data.xlsx")
df
```

```python
# Filter all the records where salary > 100000
df[df.salary > 100000]
```

```python
# Filter all records where salary > 100000 and(&) salary < 150000
df[(df.salary > 100000) & (df.salary < 150000)]
```

```python
# Query: filter all the reports where the salary < 100000 or(|) salary > 150000
df[(df.salary < 100000) | (df.salary > 150000)]
```

```python
# Query: print the record of the person who is having the highest/max salary
df[df.salary == df.salary.max()]
```

```python
# Query: print the record of the person who is having the Lowest/min salary
df[df.salary == df.salary.min()]
```

```python
# Where() function for filtering the data
# where() function is returning those records which are satisfying the condition, 
# other reports are updated as NaN
df2 = df.copy()
df2.Age = df2.Age.where(df2.Age > 50,16)
df2
```

### Indexing/Slicing using loc and iloc objects

```python
df.loc[1:4,'ID':'Age']
```

```python
df.loc[3:10,['ID',"Profession",'Age']]
```

```python
df.iloc[3:11,0:4]
```

```python
# iloc is more faster than the loc as loc in backend is converted to index to work on
# loc uses column names to slice
# iloc uses index values to slice
# iloc is more comfortable than the loc
# iloc's last index is excluded
```

```python
df.iloc[1:6,[0,1,4]]
```

```python
# combining loc & iloc with conditional statement
df.loc[df.Age > 50, 'ID':'Age']
```

```python
# Query: Print those records where Age > 50 & Age < 60
df.loc[(df.Age > 50) & (df.Age < 60),'ID':'Age']
```

```python
# Query: Print all records where age > 50 using iloc
df.iloc[(df.Age > 50).values,0:]
```

### Update the dataset using loc & iloc

```python
df1 = df.copy()
```

```python
# Query: we want to make the people young, change age of 50 - 60 years = 16
# here specifying "Age" is important because if not mentioned it will update
# to all columns as 16 where the age condition is met.
df1.loc[(df1.Age >= 50) & (df1.Age <= 60),"Age"] = 16
df1
```

```python
df1.Age.unique()
```

```python
# Updating the single value in dataset using loc
df.loc[df.ID == 500,'Name'] = 'Shreyas'
df
```

```python
df.loc[0,'Name'] = 'Prabhas'
df
```

```python
# Updating the single value in dataset using iloc
df.iloc[0,1] = "Ajith"
df
```

```python
df.at[1,'Name'] = 'Samantha'
df
```

```python
df.iat[2,1] = 'John'
df
```

```python
# sorting of dataset and changing the index of dataset
df.sort_values('Age',ascending = False).head(20) # by default is ascending
```

```python
# sorts based upon the ASCII value of the characters for the columns including strings
df.sort_values('Name',ascending = False).head(20)
```
