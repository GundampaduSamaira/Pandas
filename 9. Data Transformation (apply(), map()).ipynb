```python
import pandas as pd
```

```python
df = pd.read_excel("Excel_data.xlsx")
df
```

```python
# apply() function lets you run the custom function on our dataset
df1 = df.copy()
df2 = df.copy()
```

```python
# Query: Today the CEO of the company is very happy.
# He decided to double the salary

# without using apply() function
df2['salary'] = df2['salary'] * 2
df2
```

```python
# with the help of apply() function 
df1['salary'] = df1.salary.apply(lambda x : x * 2)
df1
```

```python
# Query: Create a new column "Category"
# 'Senior' if salary > 100000 otherwise 'Junior'
df1['Category'] = df1.apply(lambda x : 'Senior' if x['salary'] > 100000 else 'Junior', axis = 1)
df1
```
