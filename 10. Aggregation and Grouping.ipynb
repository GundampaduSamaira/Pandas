### Grouping the Data

```python
# we want to group the data based upon subclass (subgroups)
df.groupby('Profession').count()
```

```python
# based upon the profession calculating total salary
df.groupby('Profession').salary.sum()
```

```python
# Multiple aggregation functions together on a set of columns
df[['Age','salary']].agg(['mean','min','max','sum','median','count'])
```

```python
# Applying different aggregation functions for different columns
df.agg({
    'Age' : ['min','max'],
    'salary' : ['mean','count','sum']
})
```

```python
# Multipleaggregation function + groupby()
df.groupby('Profession').agg({
    'Age' : ['min','max'],
    'salary' : ['mean','count','sum']
})
```

```python
# merge(): help us to merge the dataframes based upon common column
df1 = pd.DataFrame({'ID' : [1,2,3], 'Place' : ['Ladakh','Kanyakumari','Coorg']})
df1
```

```python
df2 = pd.DataFrame({'ID' : [2,3,4], 'Best_Time_to_Visit' : ['Rainy','Winter','Winter']})
df2
```

```python
merged_df = pd.merge(df1,df2, on = 'ID') # by default merge performs inner merge
merged_df
```

```python
merged_df = pd.merge(df1, df2, on = 'ID', how = 'inner')
merged_df
```

```python
# outer merge : Return all the data of right + left table but common data will appear only once
merged_df = pd.merge(df1,df2, on = 'ID', how = 'outer')
merged_df
```

```python
# Left merge : Return all the data of left table but only common record data of right table
merged_df = pd.merge(df1,df2, on = 'ID', how = 'left')
merged_df
```

```python
# Right merge : Return all the data of right table but only common record data of left table
merged_df = pd.merge(df1,df2, on = 'ID' , how = 'right')
merged_df
```

```python
# join() : join the dataframe, based upon the Index
df1 = pd.DataFrame({'Name' : ['chill','rosy','bunny']}, index = [1,2,3])
df1
```

```python
df2 = pd.DataFrame({'Networth' : [100, 200, 560]}, index = [2,3,4])
df2
```

```python
# merge function merges based upon common column
# join is based upon common index
# default is left join
joined_df = df1.join(df2)
joined_df
```

```python
joined_df = df1.join(df2, how = 'right')
joined_df
```

```python
joined_df = df1.join(df2, how = 'left')
joined_df
```

```python
joined_df = df1.join(df2, how = 'outer')
joined_df
```

```python
joined_df = df1.join(df2, how = 'inner')
joined_df
```
