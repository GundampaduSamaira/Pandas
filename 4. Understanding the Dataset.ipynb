### Inserting/Updation new column in the dataframe

```python
df1 = df.copy()
df1['salary'] = 10000000
df1
```

```python
# syntax: dataframe_name[col_name] = value if the col_name already exist then it will update the column with value,
# but if the column does not exist then it will insert a new column with the given column name and with the assigned value.
df1["salary"] = 10000000 # update
df1["updated_salary"] = 1000 # inserted new column
df1
```

### Create a new column with rank() function

```python
# rank is given to the salary column default highest rank is given to the highest salary
# but if we want highest rank for lower salary the give ascending = False
df1['rank'] = df1.salary.rank()
df1
```

```python
df['salary_rank_ascending'] = df.salary.rank() # higher salary will get highest rank
df
```

```python
df['salary_rank_descending'] = df.salary.rank(ascending = False) # Lower salary will get highest rank
df
```

```python
df1.sort_values('rank')
```

```python
df.sort_values('salary_rank_descending')
```
