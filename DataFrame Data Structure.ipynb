```python
import pandas as pd
```

```python
df = pd.DataFrame()
print(df)
```

```python
# list into a 1D dataframe
l1 = ['pandas',45,67,'A',23.67]
df = pd.DataFrame(l1)
print(df)
print(type(df))
```

```python
# nested list of list into a 2D dataframe
l2 = [[1,2],[3,6,9],[4,7]]
df = pd.DataFrame(l2)
print(df)
```

```python
# nested tuple of tuple into a 2D dataframe
t1 = (('E011','Swetha'),('E025','Keshav'),('E045','Jake'),('E124','josh'))
df = pd.DataFrame(t1)
print(df)
```

```python
# nested list of set into a 2D dataframe
# sets cannot store mutable type of data in it.
l3 = [{1,'kitkat'},{3,'Fuse'},{2,'jimjam'},{4,'CrackJack'}]
df = pd.DataFrame(l3)
print(df)
```

```python
# create a DataFrame from the Dictionary
food_dict = {
    "Name":['Five Star','kitkat','Fuse','Crack Jack','Dairy Milk'],
    "Count":[5,50,34,12,65]
}
df = pd.DataFrame(food_dict)
print(df)
```

```python
# Convert the Series to DataFrame
ser = pd.Series([1,2,3,4,5],name = 'Count')
df1 = pd.DataFrame(ser)
df2 = ser.to_frame()
print(ser)
print(df1)
print(type(df1))
print(df2)
print(type(df2))
```
